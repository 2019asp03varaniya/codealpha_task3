```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
```

```python
car_data=pd.read_csv('car data.csv')
```

```python
car_data.head()
```

```python
car_data.isnull().sum()
```

```python
car_data['Car_Name'].unique()
```

```python
car_data['Fuel_Type'].unique()
```

```python
fuel_type={'Petrol':0,'Diesel':1,'CNG':2}
```

```python
car_data['Fuel_Type']=car_data['Fuel_Type'].map(fuel_type)
```

```python
car_data.head()
```

```python
car_data['Selling_type'].unique()
```

```python
sell={'Dealer':0,'Individual':1}
car_data['Selling_type']=car_data['Selling_type'].map(sell)
```

```python
car_data.head()
```

```python
car_data['Transmission'].unique()
```

```python
Trans={'Manual':0,'Automatic':1}
car_data['Transmission']=car_data['Transmission'].map(Trans)
```

```python
encode=LabelEncoder()
car_data['Car_Name']=encode.fit_transform(car_data['Car_Name'])
```

```python
car_data.head()
```

```python
car_data.isnull().sum()
```

```python
model=LinearRegression()
```

```python
features=['Car_Name','Year','Present_Price','Driven_kms','Fuel_Type','Selling_type','Transmission','Owner']
X=car_data[features]
y=car_data['Selling_Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
MSE=mean_squared_error(y_test,y_pred)
```

```python
print('Mean squared error',MSE)
```
